```python
# Cumulative Distributions for Feb, March

FIG_ElecCDF = plt.figure()
PLT_ElecCDF_Feb = elec_and_weather['USAGE'].ix['01-feb-2014 00:00:00':'28-feb-2014 23:00:00'].hist(normed=1,cumulative=True,histtype='step',bins=200,grid=False)
PLT_ElecCDF_Mar = elec_and_weather['USAGE'].ix['01-mar-2014 00:00:00':'31-mar-2014 23:00:00'].hist(normed=1,cumulative=True,histtype='step',bins=200,grid=False)
xlim(xmax=max(elec_and_weather['USAGE'].ix['01-mar-2014 00:00:00':'31-mar-2014 23:00:00']))
#xlim(xmax=2.5)
ylim(ymax=1)
ylabel('Probability (x <= X)')
xlabel('Usage (kWh)')
title('Elec. Usage Cumulative Distributions \n February and March 2014')
PLT_ElecCDF.text(2.3,0.072,'N = %d hours'%size(elec_and_weather['USAGE']))
legend(['Feb','March'],'lower right')

FIG_ElecCDF.savefig('ElecCDF - Feb_March2014.png')
```

```python
# Group into weekday/weekend and hour, repeat for Feb and March alone

GB_WkdayWkend_Hour = elec_and_weather.groupby([(elec_and_weather.index.dayofweek==5)|(elec_and_weather.index.dayofweek==6),elec_and_weather.index.hour])
AvgDay = pd.DataFrame([GB_WkdayWkend_Hour['USAGE'].mean(),GB_WkdayWkend_Hour['tempF'].mean()])

elec_and_weather_Feb = elec_and_weather.ix['01-feb-2014 00:00:00':'28-feb-2014 23:00:00']
GB_WkdayWkend_Hour_Feb = elec_and_weather_Feb.groupby([(elec_and_weather_Feb.index.dayofweek==5)|(elec_and_weather_Feb.index.dayofweek==6),elec_and_weather_Feb.index.hour])
AvgDay_Feb = pd.DataFrame([GB_WkdayWkend_Hour_Feb['USAGE'].mean(),GB_WkdayWkend_Hour_Feb['tempF'].mean()])

elec_and_weather_Mar = elec_and_weather.ix['01-mar-2014 00:00:00':'31-mar-2014 23:00:00']
GB_WkdayWkend_Hour_Mar = elec_and_weather_Mar.groupby([(elec_and_weather_Mar.index.dayofweek==5)|(elec_and_weather_Mar.index.dayofweek==6),elec_and_weather_Mar.index.hour])
AvgDay_Mar = pd.DataFrame([GB_WkdayWkend_Hour_Mar['USAGE'].mean(),GB_WkdayWkend_Hour_Mar['tempF'].mean()])

```
